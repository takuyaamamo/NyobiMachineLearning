### 人工データの作成

```python
# scikit-learn ライブラリから make_blobs() 関数を import 
from sklearn.datasets import make_blobs

X, y = make_blobs(n_samples=200, centers=4, n_features=2, cluster_std=2.0, random_state=42)
print(X[:10])
print(y[:10])
'''
人工データを作成。
n_samples はデータ数、centers はクラスタの数、n_features は特徴量の次元、cluster_std はデータのばらつき具合（標準偏差）を表すパラメータ
4つのクラスタを持つ特徴量が2つの200個のサンプルを作成
'''
```

### 人口データの可視化

```python
%matplotlib inline
from matplotlib import pyplot
import japanize_matplotlib

pyplot.figure(figsize=(12,12))

# データの散布図をプロットする
pyplot.scatter(X[:, 0], X[:, 1], s=100)

pyplot.xlabel(f'$x_{0}', fontsize=16)
pyplot.ylabel(f'$x_{1}', fontsize=16)
pyplot.grid()
pyplot.show()
```

### 人工データなので正解ラベルを確認できる

```python
# クラス毎にマーカーを変える
markers = ['o', 'v', 's', 'p']

pyplot.figure(figsize=(12,12))

for i in range(4):
    idx = (y == i)# クラスタ番号が一致するインデックス
    
    # 散布図をプロット
    pyplot.scatter(X[idx, 0], X[idx, 1], s=100, marker=markers[i], label=f'クラスタ{i}')
    
pyplot.xlabel(f'x_{0}', fontsize=16)
pyplot.ylabel(f'x_{1}', fontsize=16)

pyplot.grid()
pyplot.legend()
pyplot.show()
```

## K-means
- ランダムに  K  個のクラスタの中心（セントロイド） μ(1),⋯,μ(K)  を決める
- 各サンプル（データ）を最も近いセントロイドに割り当てる
- 割り当てられたサンプルの中心となるようにセントロイドを更新する
- サンプル点へのクラスタの割り当てが変化しなくなるか、または一定回数、手順2と3を繰り返す

```python
#  k-means の実装である KMeans を import
from sklearn.cluster import KMeans

K = 4

kmeans = KMeans(n_clusters=K, init='random', n_init=1, random_state=42)
'''
n_clusters=4, init='random', n_init=1 のパラメータを持たせて、KMeans のインスタンスを作成。
n_clusters=4 はクラスタ数を4
init はセントロイドの初期化方法を指定し,init='random' はデータからランダムに選択したものをセントロイドの初期値
n_init は セントロイドの初期化を何パターン試すかという回数を意味し、n_init=1 は1回ということになります。
'''
```

```python
# fit_predict() メソッドを利用して、学習と予測を同時に行う
y_pred = kmeans.fit_predict(X)
```

```python
print(y_pred)
```

```python
# クラスタ内距離平方和を確認してみます。 これは inertia_ 属性に格納されている
kmeans.inertia_
```

### 学習データのプロット

```python
pyplot.figure(figsize=(12,12))

# 予測のプロット
for i in range(K):
    idx = (y_pred == i)
    pyplot.scatter(X[idx,0], X[idx,1], s=50, label=f'cluster{i}')
    
# セントロイドのプロット
pyplot.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s=200, marker='*', label='centroids')

pyplot.xlabel(f'$x_{0}$', fontsize=16)
pyplot.ylabel(f'$x_{1}$', fontsize=16)
pyplot.grid()
pyplot.legend()
pyplot.show()
```

### n_initを調節

```python
# クラスタ数は同じ
K = 4

# n_initを10にする
kmeans = KMeans(n_clusters=K, init='random', n_init=10, random_state=42)
# 学習と予想
y_pred = kmeans.fit_predict(X)
```

```python
# クラスタ内距離平方和を確認
kmeans.inertia_
```

## データのプロット

```python
pyplot.figure(figsize=(12, 12))

for i in range(K):
    idx = (y_pred == i)
    pyplot.scatter(X[idx, 0],
                   X[idx, 1],
                   s=50,
                   label=f"cluster {i}")

pyplot.scatter(kmeans.cluster_centers_[:, 0],
               kmeans.cluster_centers_[:, 1],
               s=200,
               marker='*',
               label='centroids')
    
pyplot.xlabel(f"$x_{0}$", fontsize=16)
pyplot.ylabel(f"$x_{1}$", fontsize=16)
pyplot.grid()
pyplot.legend()
pyplot.show()
```

```python
# セントロイドの座標を確認
kmeans.cluster_centers_
```

### Kの値が3の場合

```python
# クラスタ数を3にする
K = 3
# クラスタ数以外のパラメータは同じ
kmeans = KMeans(n_clusters=3, init='random', random_state=42)
# 学習と予測
y_pred = kmeans.fit_predict(X)
```

```python
# クラスタ内距離平方和を確認します。
kmeans.inertia_
# クラスタない距離平方和は大きく悪い
```

```python
pyplot.figure(figsize=(12, 12))

for i in range(K):
    idx = (y_pred == i)
    pyplot.scatter(X[idx, 0],
                   X[idx, 1],
                   s=50,
                   label=f"cluster {i}")

pyplot.scatter(kmeans.cluster_centers_[:, 0],
               kmeans.cluster_centers_[:, 1],
               s=200,
               marker='*',
               label='centroids')
    
pyplot.xlabel(f"$x_{0}$", fontsize=16)
pyplot.ylabel(f"$x_{1}$", fontsize=16)
pyplot.grid()
pyplot.legend()
pyplot.show()
```
