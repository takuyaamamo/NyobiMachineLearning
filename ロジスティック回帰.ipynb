```python
# データの読み込みと前処理
from pandas import DataFrame
from sklearn.datasets import load_breast_cancer
```

```python
breast_cancer = load_breast_cancer()
X = breast_cancer.data[:, :10]
y = breast_cancer.target

columns = ['半径', 'テクスチャ', '周囲の長さ', '面積', 'なめらかさ', 'コンパクト性', 'へこみ', 'へこみの数', '対称性', 'フラクタル次元']

df = DataFrame(data=X[:, :10], columns=columns)
print('df', df[:10])
df['目的変数'] = y
```

```python
X = df[['面積', 'へこみ']].values
y = df['目的変数'].values
```

```python
from sklearn.model_selection import train_test_split
# 訓練データとテストデータに分ける
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler

# StandardScalerのインスタンスを作成
sc = StandardScaler()
# 訓練データの平均と標準偏差を計算する。ここでX_trainの平均と、標準偏差がscにインプットされる
sc.fit(X_train)
```

```python
# 訓練データの標準化（平均が0標準偏差が1）
X_train_std = sc.transform(X_train)
print('X_train_stdの平均値', X_train_std.mean(axis=0))
print('X_train_stdの標準偏差', X_train_std.std(axis=0))
# テストデータの標準化
# テストデータは必ず、訓練データの平均と標準偏差を用いて変換する。
X_test_std = sc.transform(X_test)
```

```python
# 学習

# 確率的勾配降下法によるロジスティック回帰の学習
from sklearn.linear_model import SGDClassifier

# SGDClassifierのインスタンス作成
Ir = SGDClassifier(loss='log', alpha=0.0001, learning_rate='constant', eta0=0.1, shuffle=True, max_iter=1000, random_state=42)
'''
loss='log' を指定することでロジスティック回帰としてモデルを作成します。
alpha はL2正則化の  λ  に相当する正則化パラメータで、デフォルト値の 0.0001 を採用しました。
learning_rate='constant' と eta0=0.1 を与えることで、確率的勾配降下法の  η  を  0.1  にセットしました。
shuffle パラメータは確率的勾配降下法においては特に重要な役割を果たすパラメータで、True を与えることで、学習の各ステップ毎に学習データをシャッフルします。これにより、ランダム性のある勾配学習を行うことができます（逆にシャッフルしないと、毎回同じ順番で勾配を計算します）。
'''
```

```python
Ir.fit(X_train_std, y_train)
```

```python
# テストデータの予測
pred = Ir.predict(X_test_std)
# 予測の確認
print(pred)
```

```python
# ロジスティック回帰は各クラスに所属する確率も得ることができます。予測確率を得るためには predict_proba() メソッドを利用
# 確率
proba = Ir.predict_proba(X_test_std)
print(proba[:10])
# クラス0（悪性）の確率が約0.125 (12.5%)、クラス1（良性）の確率が約0.875 (87.5%) であることを意味
```

```python
# 評価
from sklearn.metrics import accuracy_score

accuracy_score(y_test, pred)
```

```python
# 決定領域のプロット
import numpy
from matplotlib import pyplot
import japanize_matplotlib
from mlxtend.plotting import plot_decision_regions
```

```python
# 全てのデータをプロットするとデータが大きすぎるので制限
N = 100

# 訓練データとテストデータからN個ずつのサンプルを先頭から取ってくる
# vstack、hstack は、配列を縦（vstack）、または横（hstack）に連結
sampled_X = numpy.vstack((X_train_std[:N], X_test_std[:N]))
sampled_y = numpy.hstack((y_train[:N], y_test[:N]))
```

```python
pyplot.figure(figsize=(12, 12))
pyplot.xlabel('面積')
pyplot.ylabel('へこみ')
pyplot.title('ロジスティック回帰の決定領域')

# 決定領域のプロット
plot_decision_regions(sampled_X, sampled_y, clf=Ir, legend=2, X_highlight=X_test_std[:N])
```
